```python
# DESAFIOS
# If, Elif, Else

'''
Criar um programa que dependendo da temperatura (em celsius) do steak ele retorna
o ponto de cozimento em portugues. O usuário deverá fornecer a temperatura.

Temperaturas de cozimento
120°F ou 48°C - Rare (Selada)
130°F ou 54°C - Medium rare (Ao ponto pra mal)
140°F ou 60°C - Medium (Ao ponto)
150°F ou 65°C - Medium well (Ao ponto pra bem)
160°F ou 71°C - Well done (Bem passada)
'''

temperatura = int(input('Qual é a temperatura da carne?: '))
if temperatura >= 71:
  print('Bem passada')
elif temperatura >= 65 and temperatura <= 70:
  print('Ao ponto pra bem')
elif temperatura >= 60 and temperatura <= 64:
  print('Ao ponto')
elif temperatura >= 54 and temperatura <= 59:
  print('Ao ponto pra mal')
elif temperatura >= 48 and temperatura <= 53:
  print('Selada')
else:
  print('Cozinhar por mais alguns minutos')
```

```python
# DESAFIOS - CORREÇÃO
# If, Elif, Else

'''
Criar um programa que dependendo da temperatura (em celsius) do steak ele retorna
o ponto de cozimento em portugues. O usuário deverá fornecer a temperatura.

Temperaturas de cozimento
120°F ou 48°C - Rare (Selada)
130°F ou 54°C - Medium rare (Ao ponto pra mal)
140°F ou 60°C - Medium (Ao ponto)
150°F ou 65°C - Medium well (Ao ponto pra bem)
160°F ou 71°C - Well done (Bem passada)
'''
tem_cel = int(input('Qual é a temperatura da carne? '))

if tem_cel < 48:
  print('Cozinhar por mais alguns minutos')
elif tem_cel in range(48, 54):
  print('Selada')
elif tem_cel in range(54, 60):
  print('Ao ponto pra mal')
elif tem_cel in range(60, 65):
  print('Ao ponto')
elif tem_cel in range(65, 71):
  print('Ao ponto pra bem')
elif tem_cel >= 71:
  print('Bem passada')
```

```python
# DESAFIOS
# FUNÇÕES
'''
Criar um programa que calcula a quantidade de tinta necessaria para pintar
uma parede. O usuário deverá fornecer as seguintes informações:
1 Rendimento
2 Altura
3 Largura
O programa deve mostrar na tela a mensagem:
"Você necessita de 'x' latas de tinta.
'''
rendimento = int(input('Qual é o rendimento da lata? '))
altura = int(input('Qual é a altura da parede? '))
largura = int(input('Qual é a largura da parede? '))

def calculo_tinta():
  area = altura * largura
  total = area / rendimento
  print(f'Você precisa de {total} latas de tinta')

calculo_tinta()
```

```python
# DESAFIOS
# 'SETS'
'''
Criar um programa que gera 3 listas de acordo com a necessidade logo abaixo:

Lista1 = Funcionarios que tem carro e trabalham a noite
Lista2 = Funcionarios que tem carro e trabalham de dia
Lista3 = Funcionários que não tem carro
'''

funcionarios = ['Ana', 'Marcos', 'Alice', 'Pedro', 'sophia', 'Bruno', 'Melissa']
turno_dia = ['Ana', 'Marcos', 'Alice', 'Melissa']
turno_noite = ['Pedro', 'Sophia', 'Bruno']
tem_carro = ['Marcos', 'Alice', 'Bruno', 'Melissa']

Lista1 = set(funcionarios)
Lista2 = set(turno_dia)
Lista3 = set(turno_noite)
Lista4 = set(tem_carro)

print(Lista3 & Lista4)
print(Lista2 & Lista4)
print(Lista1 ^ Lista4)
```

```python
# DESAFIOS
# CALCULO DE IMC - Indice de Massa Corporal
'''
Qual é a sua altura em cm:
Qual é o seu peso em Kg
'''
# Menor que 18,5     = MAGREZA
# Entre 18,5 e 24,9  = NORMAL
# Entre 25,0 e 29,9  = SOBREPESO
# Entre 30,0 e 39,9  = OBESIDADE
# Maior que 40,0     = OBESIDADE GRAVE

altura = float(input('Qual é a sua altura em cm?: '))
peso = float(input('Qual é o seu peso em Kg?: '))
imc = peso / ((altura/100) * (altura/100))
print(imc)
if  imc >= 40.0:
  print('OBESIDADE GRAVE')
elif imc >= 30.0 and imc <= 39.9:
  print('OBESIDADE')
elif imc >= 25.0 and imc <= 29.9:
  print('SOBREPESO')
elif imc >= 18.5 and imc <= 24.9:
  print('NORMAL')
else:
  print('MAGREZA')
```

```python
# Functions (Funções)
  # DRY - Don't Repeat Yourself.
  # Vários argumentos {xargs ( * )}
# Criar uma função que soma vários números.

def soma(*numeros):
  resultado = 0
  for num in numeros:
    resultado += num
  return resultado


x= soma(2,3,4,7)

print(x)

# for (é de for loop) - significa que roda várias vezes somando os resultado
# com os próximos números
# EXEMPLO: 0+2=2  2+3=5  5+4=9  9+7=16
```

```python
# Functions (Funções)
  # DRY - Don't Repeat Yourself.
  # Vários argumentos {xargs ( * )} identificando o parâmetro ( * ) depois do outro *

# Criar uma função que armazena numeros e strings (dados)

def agencia(**carro):
  return carro

print(agencia(marca='gol', cor='branca', motor=1.8, placa=1234))
print(agencia(marca='palio', cor='verde', placa=5678))
print(agencia(marca='monza', motor=2.0, placa=9876))

# Observação: quando se utiliza dois asteríscos **, significa que:
# pode-se ter vários argumentos e vários parâmentro
# e não é necessário ter as mesmas informações em todas as "linhas"
```

```python
# Set (Listas)
  # Similar a listas
  # Evita itens duplicados
  # Não utiliza index

list1 = [10, 20, 30, 40, 50]
list2 = [10, 20, 60, 70]

num1 = set(list1)
num2 = set(list2)

print(num1 | num2) # Union - retira os valores duplicados, deixando apenas 1
print(num1 - num2) # Difference - Mostra o valor que sobra do num1, retirando os duplicados
print(num1 ^ num2) # Symmetric Difference - retira todos duplicados, deixando apenas os que não são
print(num1 & num2) # And - indica o que tem de duplicado nas listas

print(len(num1)) # indica a quantidade de itens que tem na lista
```

```python
# Lambda Function:
  # É uma função (pequena) sem nome
  # Pode ter vários argumentos mas somente 1 expressão
  # Muito utilizada dentro de outras funções
  # Código mais 'clean'

def somar(x):
  return x + 10

print(somar(2))

somar1 = lambda x: x + 10
print(somar1(2))

somar2 = lambda x,y: x + y + 10
print(somar2(2,4))
```

```python
from sys import getsizeof

# Generator Expressions
  # Uma forma mais rápida para listas, dicionários e etc.
  # Menos memória alocada
  # Valores em bytes

numeros = [x * 10 for x in range(10)]
print(type(numeros))
print(numeros)
print(getsizeof(numeros))

print('============')

numeros = (x * 10 for x in range(10))
print(type(numeros))
print(list(numeros))
print(getsizeof(numeros))

'''
Para transformar uma lista em um generator expressions, é preciso troacar os
  colchetes por parenteses
'''
```

```python
from datetime import datetime

# Python Pbject-Oriented Programming ( Programação Orientada por Objetos)

# Classes
  # Utilizadas para criar objetos (instances)
  # Objetos são partes dentro de uma "Class" (instancias)
  # Classes são utilizadas para agrupar dados e funções, podendo reutilizar:
  # ====
  # Class: Frutas
  # Objects: Abacate, Banana, ...

# criar classe
class Funcionarios:

  def __init__(self, nome, sobrenome, ano_nascimento): # 2 underlines
    self.nome = nome
    self.sobrenome = sobrenome
    self.ano_nascimento = ano_nascimento
  
  def nome_completo(self):
    return (self.nome + ' ' + self.sobrenome)

  def idade_funcionario(self):
    ano_atual = datetime.now().year
    self.ano_nascimento = int(ano_atual - self.ano_nascimento)
    return self.ano_nascimento

# criar o objeto
usuario1 = Funcionarios('Elena', 'Cabral', 2009)
usuario2 = Funcionarios('Carol', 'Silva', 2005)
usuario3 = Funcionarios('Andre', 'Iacono', 2003)

# print
print(Funcionarios.idade_funcionario(usuario1))
print(Funcionarios.idade_funcionario(usuario2))
print(Funcionarios.idade_funcionario(usuario3))
```
